```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.draw(output='mpl')
```

```python
from qiskit.quantum_info import Pauli
ZZ=Pauli('ZZ')
ZI=Pauli('ZI')
IZ=Pauli('IZ')
XX=Pauli('XX')
IX=Pauli('IX')

OBSV=[ZZ,ZI,IZ,XX,IX]
```

```python
from qiskit_aer.primitives import Estimator
estimator=Estimator()
R=estimator.run([qc]*len(OBSV),OBSV)
R.result()
```

```python
import matplotlib.pyplot as plt
data=['ZZ','ZI','IZ','XX','IX']
values=R.result().values
plt.plot(data,values,'-o')
plt.xlabel('Obesrvables')
plt.ylabel('Expectation Value')
plt.show()
```
